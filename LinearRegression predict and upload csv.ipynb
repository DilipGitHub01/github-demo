```python
df = pd.read_csv("gapminder.csv")
df
```

```python
df.head(3)
```

```python
%matplotlib inline
plt.xlabel('Fertility')
plt.ylabel('Life')
plt.scatter(df.fertility,df.life,color = 'red',marker='+')
```

```python
X = (df[df.columns[[0,1,2,3,4,5,6,8]]].values)
y = df.life
X.shape
```

```python
reg = linear_model.LinearRegression(fit_intercept = True, normalize = True)
reg
```

```python
reg.fit(X,y)
```

```python
d = pd.read_csv('gapminder_prediction.csv')
d
```

```python
p =reg.predict(d)
```

```python
d['life'] =p
```

```python
d.to_csv('prediction_gapminer.csv')
```

```python
%matplotlib inline
plt.xlabel('Fertility')
plt.ylabel('Life')
plt.scatter(df.fertility,df.life,color = 'red',marker='+')
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3, random_state=5)
```

```python
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
print(r2_score(y_test,y_pred))
```

```python
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
print(rmse)
```

```python
std = np.std(y_test)
print(std)
```

```python
rmse<std
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cvs_score = cross_val_score(reg,X,y,cv=5)
```

```python
print(np.mean(cvs_score))
```
